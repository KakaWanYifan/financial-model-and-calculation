```python
# 第二问

# 导入 pandas，用以读取excel
import pandas as pd
# 导入 matplotlib.pyplot，用以画图
from matplotlib import pyplot as plt
df = pd.read_excel('IBM.xlsx')
x = df[['日期']].values
y = df[['收盘价']].values

plt.plot(x,y)
plt.show()
```

```python
# 第三问

import math

df['简单收益率'] = df.apply(lambda x: (x['收盘价'] / x['开盘价']) - 1, axis=1)
df['对数收益率'] = df.apply(lambda x: math.log((x['收盘价'] / x['开盘价']),math.e), axis=1)
print(df)
```

```python
# 第四问

x = df[['日期']].values
y = df[['对数收益率']].values

plt.plot(x,y)
plt.show()
```
